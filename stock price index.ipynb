```python
def plot_closing_price(df):
    """
    Plot the closing price of a DataFrame across the index.

    Parameters:
    - df: DataFrame with a DatetimeIndex and a 'Close' column.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['Close'], label='Closing Price', color='blue')
    plt.title('Closing Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def predict(train, test, predictors, model):
     model.fit(train[predictors], train["Target"])
     preds= model. predict(test[predictors])
     preds=pd.Series(preds, index= test.index, name="predictions")
     combined = pd.concat([test["Target"], preds], axis=1)
     return combined
              
```

```python
def backtest(data, model, predictors, start=0, step=100):
    all_predictions = []  # Initialize an empty list to store predictions
    for i in range(start, len(data)-step, step):
        train = data.iloc[i:i+1].copy()
        test = data.iloc[i:i+step].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions=backtest(sp500, model, predictors)
```

```python
horizons = [2, 5, 60, 100, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500['Close'].rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500["Close"].shift(1).rolling(horizon).sum()
    
    new_predictors += [ratio_column, trend_column]
```
